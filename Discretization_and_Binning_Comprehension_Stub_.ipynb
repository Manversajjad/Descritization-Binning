## Problem Statement

Let us go back to the job fair problem. As of now, you have cleaned the data and removed the duplicate entries.


Your task now is to discretize the values of the column etest_p into three quantiles, ['Low', 'Average', 'High'], and store the result in a new column, etest_c.


Transform the data based on the above conditions and answer these questions.

```python
#Importing the library
import numpy as np
import pandas as pd

#Loading the data
hiring_data = pd.read_csv('/content/hiring_data.csv')
```

Perform the previous steps

```python
#removing the rows that have more than four missing values
subset_df = hiring_data.dropna(thresh=5)
round(subset_df.salary.mean(),2)

#dropping duplicates
subset_df.drop_duplicates(inplace=True)
subset_df

#mapping values
map_label = {'S&T': 'Sci&Tech', 'C&M': 'Comm&Mgmt'}
subset_df.replace(map_label,inplace = True)
subset_df.head()
```

Use this data to answer the following questions

**Q1:**  Q1. How many students scored ‘High’ on the Employability test? [Hint: Use pd.qcut() to categorize the age groups into three quantiles.]


a)   33

b)   32

c)   31

d)   30





```python
# Dropping rows with missing 'etest_p' values to ensure accurate quantile calculation
hiring_data_cleaned = hiring_data.dropna(subset=['etest_p'])

# Categorizing 'etest_p' into three quantiles: Low, Medium, and High
hiring_data_cleaned['etest_category'] = pd.qcut(hiring_data_cleaned['etest_p'],
                                                q=3,
                                                labels=['Low', 'Medium', 'High'])

# Counting the number of students in the 'High' category
high_category_count = hiring_data_cleaned[hiring_data_cleaned['etest_category'] == 'High'].shape[0]

high_category_count


```

```python
# Filtering students who scored "High" on the employability test
High_scorers = hiring_data_cleaned[(hiring_data_cleaned['etest_category'] == 'High')]
# To calculate the average
Average_salary = High_scorers['salary'].mean()
# Round to 2 decimal places
Rounded_val_ofDecimals = round(Average_salary, 2)
Rounded_val_ofDecimals
```
