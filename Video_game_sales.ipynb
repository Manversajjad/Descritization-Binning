```python
import numpy as np
import pandas as pd
```

```python
vd = pd.read_csv(r'/content/video-game-sales.csv')
vd.head(5)
```

**Changing the RANK Index to match with the numbering of un-numbered Index**

```python
vd.index = vd['Rank']
vd.head()
```

Delete the Index RANK value to clean the data to perform the analysis unto.

```python
del vd['Rank']
vd.head()
```

Explore the Distribution and Insights from the dataset

```python
vd['Year'].describe()
```

Create Linearly Spaced Bins. To Analyze The Data Every 5 Year

```python
bins_linear = np.linspace(start=1975, stop=2020, num=10)
bins_linear
```

Bin the Year values based on previously created bins. When we cut the value, First Number or year will be excluded and will start from the next year onwards

```python
vd['Year Bins'] = pd.cut(vd['Year'], bins_linear)
vd[['Name', 'Platform', 'Year', 'Year Bins']].head(10)
```

If wanted to see the number of counts in each Bins. We can use value_counts() function

```python
vd['Year Bins'].value_counts().sort_index()
```

Quick Descretization Data

```python
vd['NA_Sales'].describe()
```

```python
vd[vd['NA_Sales'] != 0]
```

Summarize the frequency of quartile Bins

```python
pd.qcut(vd[vd['NA_Sales'] != 0]['NA_Sales'], 4).value_counts().sort_index()
```

Summarize the frequency of decile bins, for north american sales of video games that have had sales. Keep in mind these values are defined to be in millions

```python
pd.qcut(vd[vd['NA_Sales'] != 0]['NA_Sales'], 10).value_counts().sort_index()
```

Add Labels to help explain quantiles to end-users

```python
Bins_label = ['Bronze', 'Silver', 'Gold', 'Platinum']
vd['Achievement_Level'] = pd.qcut(vd[vd['NA_Sales'] != 0]['NA_Sales'],q=4, labels=Bins_label)
vd.head(10)
```
